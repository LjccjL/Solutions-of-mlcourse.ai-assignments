```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

**Read data**

```python
train_df = pd.read_csv("../../data/titanic_train.csv", 
                       index_col='PassengerId') 
```

```python
train_df.head(2)
```

```python
train_df.describe(include='all')
```

```python
train_df.info()
```

**Let's drop`Cabin`, and then â€“ all rows with missing values.**

```python
train_df = train_df.drop('Cabin', axis=1).dropna()
```

```python
train_df.shape
```

**1. Build a picture to visualize all scatter plots for each pair of features `Age`, `Fare`, `SibSp`, `Parch` and `Survived`. ( `scatter_matrix ` from Pandas or `pairplot` from Seaborn)**

```python
train_df[['Age','Fare','SibSp','Parch','Survived']].head(2)
```

```python
pd.plotting.scatter_matrix(train_df[['Age','Fare','SibSp','Parch','Survived']],
                           figsize=(12,10));
```

```python
sns.pairplot(train_df[['Age','Fare','SibSp','Parch','Survived']]);
```

**2. How does ticket price (`Fare`) depend on `Pclass`? Build a boxplot.**

```python
pd.plotting.boxplot(data=train_df,column=['Fare'],by='Pclass');
```

```python
sns.boxplot(x='Pclass',y='Fare',data=train_df);
```

**3. Let's build the same plot but restricting values of `Fare` to be less than 95% quantile of the initial vector (to drop outliers that make the plot less clear).**

```python
train_df['Fare'].quantile(q=0.95)
```

```python
sns.boxplot(x='Pclass',y='Fare',data=train_df[train_df['Fare']<=120.0]);
```

**4. How the percentage of survived passengers is dependent on passengers' gender? Depict it with `Seaborn.countplot` using the `hue` argument.**

```python
sns.countplot(x='Survived',hue='Sex',data=train_df);
```

**5. How does the distibution of ticket prices differ for those who survived and those who didn't. Depict it with `Seaborn.boxplot`**

```python
sns.boxplot(x='Survived',y='Fare',data=train_df);
```

```python
sns.boxplot(x='Survived',y='Fare',data=train_df[train_df['Fare']<=120]);
```

**6. How survival depended on passengers' age?  Verify (graphically) an assumption that youngsters (< 30 y.o.) survived more frequently than old people (> 55 y.o.).**

```python
sns.countplot(x='Survived',data=train_df[train_df['Age']<30]);
```

```python
sns.countplot(x='Survived',data=train_df[train_df['Age']>55]);
```

```python
# A better way
train_df['Age_30/55']=train_df['Age'].apply(lambda age: 1 if age<30  
                                            else 3 if age>55       
                                            else 2);
```

```python
pd.crosstab(train_df['Age_30/55'],train_df['Survived'],margins=True)
```

```python
sns.countplot(x='Age_30/55',hue='Survived',data=train_df);
```
